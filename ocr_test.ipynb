```python
import easyocr
```

```python
reader = easyocr.Reader(['ko', 'en'], gpu=True)
txt = reader.readtext('test.jpg')
print(txt)
```

```python
import cv2
img = cv2.imread(filename='test.jpg')
img_ = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
grayAndBlur = cv2.GaussianBlur(img_, (3, 3), 0)
```

```python
textBox_gray = reader.detect(grayAndBlur, 30, width_ths=2, height_ths=1)
```

```python

color = (255, 0, 0)
for t in textBox_gray[0][0]:
    # print(t)
    topLeft = [t[0], t[2]]
    bottomRight = [t[1], t[3]]
    # print(topLeft, bottomRight)
    cv2.rectangle(img, topLeft, bottomRight, color)
```

```python
# img = cv2.cvtColor(img, cv2.COLOR_BGRA2RGB)
cv2.imshow("gray", img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
