```python
img = cv2.imread(filename='test.jpg')
print(img.shape)        # height, width, channel
ratio = img.shape[0]/img.shape[1]
height = 1080*ratio
img_ = cv2.resize(img, dsize=(1080, int(height)))
grayAndBlur = cv2.GaussianBlur(img_, (3, 3), 0)
print(img_.shape)
```

```python
textBox_gray = reader.detect(grayAndBlur, 30, text_threshold=0.5, width_ths=1, height_ths=1)
```

```python

color = (255, 0, 0)
for t in textBox_gray[0][0]:
    # print(t)
    topLeft = [t[0], t[2]]
    bottomRight = [t[1], t[3]]
    # print(topLeft, bottomRight)
    cv2.rectangle(img_, topLeft, bottomRight, color)
```
